# Meditation #2: Machine Learning or "Machine Rummaging"?

---

#### *I didn't go gleaning in the forest to find something in particular; I went to find whatever there was to find.* - Uprooted, **Naomi Novik**

Like a gleaner wandering through a forest, a model does not set out with a hypothesis or destination — it gathers *whatever patterns there are to find* in its training distribution. Some of these patterns are sturdy and meaningful; others are coincidences, biases or statistical debris. But the process is often the same: **opportunistic pattern-collection without intent**.

This framing matters.

We call the field “machine learning”, but to those outside or new to the field may be tempted to imagine comprehension, judgement or purpose where none exists. But quitely internalising it as **machine rummaging** encourages the right mental model: the model is an extraordinarily efficient gleaner, not a thinker. It will retrieve correlations whether or not they make sense — and it will do so with complete indifference to your goals.

This is where **ANCD** (<u>link coming!</u>) becomes useful: the model’s tendency to pull everything toward an approximate nearest common denominator — collapsing variety into whatever regularities dominate the data — is a natural outcome of rummaging. The machine doesn’t “understand” why certain structures recur; it merely notices and amplifies them, even when they’re fragile, spurious or misleading.

**The responsibility, therefore, lives with us.**

A rummager can gather; only a human can interpret. Our job is to examine what the model has dragged back from the forest — test it, stress it, contextualise it — and decide which patterns merit trust and which should be returned to the undergrowth.

This is perhaps what all models should have appended to their output:
> #### From what I know of the world — meaning the data you’ve trained me on — these are the patterns I’ve discovered. Some may be coincidence or nonsense. Over to you to validate with metrics, and use your subject expertise to decide what (if anything) they mean.
