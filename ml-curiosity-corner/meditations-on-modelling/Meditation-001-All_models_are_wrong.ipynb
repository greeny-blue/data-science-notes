# Meditation #1: All Models Are Wrong, But Some Are Useful

---

[George E. P. Box](https://en.wikipedia.org/wiki/George_E._P._Box) once stated that ["all models are wrong"](https://en.wikipedia.org/wiki/All_models_are_wrong). It's something I think about whenever I come to modelling data. 

His point was simple but profound: a **model** is a **representation**, *not the thing itself*. This has been true long before algorithms and datasets $-$ it’s just easier to forget now that our models are more complex.

When someone shapes clay into a figure, it is already a model $—$ it captures certain **features** and leaves countless details out. A map is a 2-D model of a 3-D world: lines where roads should be, symbols standing in for buildings, elevation reduced to contour marks. Some **dimensions** are kept, others deliberately suppressed.

Every act of modelling involves **selection**. A sketch of a tree is a model and can take many forms: emphasis on light and shadow, or colour and gesture, with countless subtleties omitted. The world is compressed into a form that serves a purpose.

Data science does the same, only with numbers. We define features, impose structure and choose what matters. The representations are more formal, but the logic is identical: a model is always an incomplete version of reality.

Recognising this doesn’t weaken modelling $—$ it clarifies it. It reframes the real question from “Is this model right?” to:

> #### “Given the way I’ve chosen to represent this thing, in what ways is the model inevitably wrong $—$ and how much does that matter for what I’m trying to do?”

This viewpoint keeps the work honest, whether you’re sculpting clay, drawing a map or training a classifier. It’s not pessimism $—$ it’s awareness.

---

*Note:* All the terms in **bold** $-$ model, representation, feature, dimension, selection $—$ are common in data science because, in most cases, they are direct analogues of their original physical meanings.
